```python
import pandas as pd
import os
```

```python
# Define the directory where the CSV files are located
directory = r"C:\Users\raulo\Desktop\CÃ³digoTese\Labels"

# Initialize an empty list to store the summary rows
summary_background = []
summary_IntClient = []
summary_IntCleaner = []
summary_Dance = []
summary_TS = []
summary_Pousing = []
summary_Jolt = []
summary_Run = []


# Iterate over the files in the directory
for file in os.listdir(directory):
    # Construct the full path to the file
    file_path = os.path.join(directory, file)


    # Load the dataset from the CSV file
    df = pd.read_csv(file_path)

    # Initialize variables to keep track of the current classification and frame range
    current_classification = ''
    start_frame = 0
    last_classification = ''


    # Iterate over the rows of the dataset
    for index, row in df.iterrows():

        # Get the name of the column with the non-zero value
        classification_column = row[1:].idxmax()

        # Check if the classification has changed or if it's the last row of the dataset
        if (row[1:] != current_classification).any() or index == len(df) - 1:

            # Get the end frame of the range
            end_frame = index - 1 if index != len(df) - 1 else index

            # Add a new summary row to the list
            if last_classification == "background":
                summary_background.append([file,start_frame, end_frame, last_classification])          
            if last_classification == "Int_Client":
                summary_IntClient.append([file,start_frame, end_frame, last_classification])
            if last_classification == "Int_Cleaner":
                summary_Int_Cleaner.append([file,start_frame, end_frame, last_classification])
            if last_classification == "Dance":
                summary_Dance.append([file,start_frame, end_frame, last_classification])
            if last_classification == "TS":
                summary_TS.append([file,start_frame, end_frame, last_classification])
            if last_classification == "Pousing":
                summary_Pousing.append([file,start_frame, end_frame, last_classification])
            if last_classification == "Jolt":
                summary_Jolt.append([file,start_frame, end_frame, last_classification])  
            if last_classification == "Run":
                summary_Run.append([file,start_frame, end_frame, last_classification])  

            # Update the variables for the next range
            current_classification = row[1:]
            start_frame = index
            last_classification = current_classification.idxmax()

    # Create a new DataFrame from the summary rows and save it to a CSV file
    summary_df_background = pd.DataFrame(summary_background, columns=['file','start_frame', 'end_frame', 'classification'])
    summary_df_IntClient = pd.DataFrame(summary_IntClient, columns=['file','start_frame', 'end_frame', 'classification'])
    summary_df_IntCleaner = pd.DataFrame(summary_IntCleaner, columns=['file','start_frame', 'end_frame', 'classification'])
    summary_df_Dance = pd.DataFrame(summary_Dance, columns=['file','start_frame', 'end_frame', 'classification'])
    summary_df_TS = pd.DataFrame(summary_TS, columns=['file','start_frame', 'end_frame', 'classification'])
    summary_df_Pousing = pd.DataFrame(summary_Pousing, columns=['file','start_frame', 'end_frame', 'classification'])
    summary_df_Jolt = pd.DataFrame(summary_Jolt, columns=['file','start_frame', 'end_frame', 'classification'])
    summary_df_Run = pd.DataFrame(summary_Run, columns=['file','start_frame', 'end_frame', 'classification'])

    summary_df_background.to_csv('summary_df_background.csv', index=False)
    summary_df_IntClient.to_csv('summary_df_IntClient.csv', index=False)
    summary_df_IntCleaner.to_csv('summary_df_IntCleaner.csv', index=False)
    summary_df_Dance.to_csv('summary_df_Dance.csv', index=False)
    summary_df_TS.to_csv('summary_df_TS.csv', index=False)
    summary_df_Pousing.to_csv('summary_df_Pousing.csv', index=False)
    summary_df_Jolt.to_csv('summary_df_Jolt.csv', index=False)
    summary_df_Run.to_csv('summary_df_Run.csv', index=False)
```
